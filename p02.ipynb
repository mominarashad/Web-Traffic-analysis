```python
#calculate all event
#creating function

def get_statistics(event_name):
  unique_event=df[df['event']==event_name]
  print('Stat for event:',event_name)
  unique_count=unique_event.groupby(by="date")['event'].count()
  print(unique_count)
```

```python
#get all events stats
for i in events:
  get_statistics(i)
```

#Country with highest click rate

```python
#actions to follow
#filter click events
#remove dublicates
#group by country and count click event
#find the country with maximum value of count
```

```python
#filter click events
cr=df[df['event']=='click']
#remove dublicates
crd=cr.drop_duplicates()
#group by country and count click event
country_count=crd.groupby("country")['event'].count()
#find the country with maximum value of count
country_count.sort_values(ascending=False).index[0]

```

#Click rate means click/pageviews

```python
#seprating the data for display link (page view event)
lv=df[df['event']=='pageview']
#group the link id and count the event
lv_c=lv.groupby(by="linkid")['event'].count()
lv_c
#converting sereies into dataframe
lv_c=lv_c.reset_index()
#renaming the column to link_view for better understanding
lv_c.rename(columns={'event':'link_view'},inplace=True)
lv_c.head()
```

#For each click count how many time it has been clicked

```python
#seprating data for click event
df_cc=df[df['event']=='click']
#grouo by link id and count the event
df_count=df_cc.groupby(by="linkid")['event'].count()
#converting series into data frame
df_count=df_count.reset_index()
#renaming the column
df_count.rename(columns={'event':'click_count'},inplace=True)
df_count.head()
```

#Merge two data frame as above to one based on linkid

```python
df_qq=pd.merge(right=lv_c,left=df_count, on="linkid",how="inner")
df_qq.head()
```

```python
df_qq['click_rate']=df_qq['click_count']/ df_qq['link_view']

#if you want to calculate percentage
df_qq['click_rate']=df_qq['click_rate']*100
df_qq.head(10)
```
